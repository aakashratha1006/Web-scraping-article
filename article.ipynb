```python
# Reading webpage into python
import requests
r = requests.get('https://www.nytimes.com/interactive/2017/06/23/opinion/trumps-lies.html')
```

```python
print(r.text[0:500]) # First 500 characters of HTML
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text,'html.parser')
soup
```

```python
results = soup.find_all('span',attrs={'class':'short-desc'})
```

BUILDING DATASET..............

```python
records = []
for result in results:
    date        = first_result.find('strong').text[0:-1] + ', 2017'
    lie         = first_result.contents[1][1:-2]
    explanation = first_result.find('a').text[1:-1]
    url         = first_result.find('a')['href']
    records.append((date, lie, explanation, url))
```

```python
len(records)
```

```python
records[0:3]
```

APPLYING A TABULAR DATA STRUCTURE..............

```python
import pandas as pd
df = pd.DataFrame(records, columns=['date', 'lie', 'explanation', 'url'])
```

```python
df.head()
```

```python
df.tail()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.to_csv('article.csv', index=False, encoding='UTF-8')
```
